```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_flights = sns.load_dataset("flights")

print(df_flights.head(20))
```

```python
print(df_flights.columns.to_list())
```

```python
sns.lineplot(data=df_flights, x="year", y="passengers", estimator="sum")
plt.title("Total Passengers per Year")
plt.show()
```

```python
# Pivot the data for heatmap
df_flights_pivot = df_flights.pivot(index="month", columns="year", values="passengers")

# Plot heatmap
sns.heatmap(df_flights_pivot, annot=True, fmt="d", cmap="magma")
plt.title("Monthly Airline Passengers (Heatmap)")
plt.show()
```

```python
df_exercise = sns.load_dataset("exercise")
print(df_exercise.head(20))
```

```python
print(df_exercise.columns.to_list())
```

```python
sns.boxplot(data=df_exercise, x="kind", y="pulse", hue="diet")
plt.title("Pulse Rate by Exercise Type with Diet")
plt.show()
```

```python
print("- Rest has the lowest and tightest range of pulse rates (~80100).\n- Walking increases the pulse moderately (~85110), with a slightly wider spread.\n- Running causes the highest variability and maximum pulse rates, peaking around 150.")
```

```python
sns.violinplot(data=df_exercise, x="kind", y="pulse", hue="diet", split=True)
plt.title("Pulse Rate by Exercise Type With Diet and Split (Violin)")
plt.show()
```

```python
print("Individuals on a low fat diet may exhibit slightly more stable cardiovascular \nresponse during intense activity compared to those on a no fat diet, \nthough more data would be needed for real conclusions.")
```
