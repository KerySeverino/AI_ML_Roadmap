```python
import pandas as pd
import seaborn as sns
from IPython.display import display


df = sns.load_dataset("titanic")
display(df)
```

```python
print("Shape of dataset: ", df.shape)
print("Columns: ", df.columns.to_list())
```

```python
avg_age = df["age"].mean()
print("Average age of passengers: ", round(avg_age, 2))
```

```python
females = df[df["sex"] == "female"]
males = df[df["sex"] == "male"]

display(females)
```

```python
display(males)
```

```python
print("Total survivors: ",(df["survived"] == 1).sum())
```

```python
print("Average fare of passengers:\n")
print(df.groupby("class", observed=False)["fare"].mean())
```

```python
print("Average survival rate by sex:\n")
print(df.groupby("sex", observed=False)["survived"].mean())

print("\nThis data reflects the real-life quote:")
print("\"Women and children first\", A policy that was followed during the Titanic evacuation.")
```

```python
print(df.isnull().sum())
```

```python
print("How many passengers have missing ages: ", df["age"].isnull().sum())
```

```python
df_no_missing_age = df.dropna(subset=["age"])
display(df_no_missing_age)

print(f"Original Database: {df.shape[0]} rows -> After dropping missing ages: {df_no_missing_age.shape[0]} rows")
```

```python
df_sorted_by_age = df_no_missing_age.sort_values("age")
display(df_sorted_by_age)
```

```python
women_first_class_survived = df[
                                (df["sex"] == "female") &
                                ((df["class"] == "First")) &
                                ((df["survived"] == 1))
                            ]
women_first_class_died = df[
                                (df["sex"] == "female") &
                                ((df["class"] == "First")) &
                                ((df["survived"] == 0))
                            ]
```

```python
display(women_first_class_survived[["sex", "class", "survived", "age", "fare"]])
```

```python
display(women_first_class_died[["sex", "class", "survived", "age", "fare"]])
```

```python
young_man_third_class_survived = df[
                                (df["sex"] == "male") &
                                (df["class"] == "Third") &
                                (df["survived"] == 1) &
                                (df["age"] < 30)
                            ]
young_man_third_class_died = df[
                                (df["sex"] == "male") &
                                (df["class"] == "Third") &
                                (df["survived"] == 0) &
                                (df["age"] < 30)
                            ]
```

```python
display(young_man_third_class_survived[["sex", "class", "survived", "age", "fare"]])
```

```python
display(young_man_third_class_died[["sex", "class", "survived", "age", "fare"]])
```

```python
fare_age_by_class = df.groupby("class", observed=False)[["age", "fare"]].agg(["mean", "count"])
display(fare_age_by_class)
```

```python
survival_by_class_sex = df.groupby(["class", "sex"], observed=False)["survived"].mean().reset_index()
display(survival_by_class_sex)

print("From the table:")
print("- Rows 0 to 1: About 96% of First Class females survived, while only 36% of First Class males did.")
print("- Rows 2 to 3: Around 92% of Second Class females survived, compared to 15% of Second Class males.")
print("- Rows 4 to 5: Approximately 50% of Third Class females survived, but just 13% of Third Class males.")

print("\nThis shows a clear trend, the lower the passenger's class, the lower their chances of survival.")
```

```python
print("Passengers by class:\n")
print(df["class"].value_counts())
```

```python
print("Passengers class distribution (percent):\n")
print(df["class"].value_counts(normalize=True) * 100)
```

```python
print("Passengers that were alone vs not alone:\n")
print(df["alone"].value_counts())
```

```python
avg_fare = df["fare"].mean()

def classify_fare(fare):
    if pd.isnull(fare):
        return "unknown"
    elif fare >= avg_fare:
        return "high"
    else:
        return "low"

df["fare_level"] = df["fare"].apply(classify_fare)
```
